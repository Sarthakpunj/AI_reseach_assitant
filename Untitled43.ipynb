```python
%pip install -U langchain-google-genai
from pydantic import BaseModel
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
from google.colab import userdata
api= userdata.get('gemini')

llm = ChatGoogleGenerativeAI(google_api_key=api,model="gemini-2.0-flash")
response = llm.invoke("what is meaning of life?")
print(response)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import PydanticOutputParser
```

```python
#creating fields that i want my llm to generate as output

class Research(BaseModel):
  topic: str
  summary: str
  sources: list[str]
  tools_used: list[str]

parser = PydanticOutputParser(pydantic_object=Research) #take the output generated by llm and pass it to the reseach class
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            """"
            you are a research assitant that will help generate a research paper.
            answer the user query and use necessary tools to generate a research paper.
            Wrap the output in this format and provide no other text\n{format_instructions}
            """,
        ),
        ("placeholder","{chat_history}"),
        ("human","{query}"),
        ("placeholder","{agent_scratchpad}")
    ]
).partial(format_instructions=parser.get_format_instructions())
```

```python
from langchain.agents import create_tool_calling_agent,AgentExecutor
```

```python
tools = [search_tools,wiki_tool]
agent = create_tool_calling_agent(
    llm = llm,
    prompt = prompt,
    tools=tools
)
```

```python
agent_exc= AgentExecutor(agent=agent,tools=tools,verbose=True)
query = input("what can i help you reserach?")
raw_res = agent_exc.invoke({"query": query})
print(raw_res)
```

```python
# now if we want response like research so we do
struct_rep = parser.parse(raw_res.get("output"))
print(struct_rep)
```

```python
from langchain_community.tools import WikipediaQueryRun,DuckDuckGoSearchRun
from langchain_community.utilities import WikipediaAPIWrapper
from langchain.tools import Tool
from datetime import datetime
```

```python
search = DuckDuckGoSearchRun()
search_tools = Tool(
    name = "search",
    func = search.run,
    description = "Search the web for information",
)
```

```python
!pip install wikipedia
api_wrapper = WikipediaAPIWrapper(top_k_results=1,doc_content_chars_max=100)
wiki_tool = WikipediaQueryRun(api_wrapper=api_wrapper)
```
